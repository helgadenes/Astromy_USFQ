## 1. Inline plotting:

```python
# Import third-party libraries
import numpy as np
import matplotlib.pyplot as plt
```

```python
# This line is added for inline, interactive plotting
%matplotlib inline

# Generate a couple of vectors
x = np.arange(0, 6*np.pi, 0.01)

y = np.sin(x)

# Plotting
plt.plot(x,y)
```

## 2. Figure environments:

- This refers to code lines that open, plot and show the figure.

```python
# The basic syntax for a fig. env.

# Open env.
plt.figure()

# Plot
plt.plot(x,y)

# Show/Save
plt.show()
# plt.savefig()

# In actual code, we should always clean up the memory
# plt.close()
```

```python
# Customise our figures:

# Open env.
plt.figure(figsize=(4,3))

# Plot
plt.plot(x, y, color = "red", linestyle = ":")

# Title with latex syntax
plt.title(r"$\sin(x)$")

# Axes titles
plt.xlabel(r"$x$")
plt.ylabel(r"$y=\sin(x)$")

# Show/Save
plt.show()
```

## Example: overplotting


```python
# Superimpose plots:

w = np.cos(x)

# Open env.
plt.figure(figsize=(4,3))

# Plot
plt.plot(x, y, color = "red", linestyle = ":", label = r"$\sin(x)$")
plt.plot(x, w, color = "green", linestyle = "-.", label = r"$\cos(x)$")

# Title with latex syntax
plt.title(r"$\sin(x)$ and $\cos(x)$")

# Axes titles
plt.xlabel(r"$x$")
plt.ylabel(r"$y,w$")

# Legend is added with loc for location
plt.legend(loc=2) # loc -> LOCATION

# Show/Save
plt.show()
```

## Example: subplots (multi-panel figures)

```python
# Method 1: Multi-panel figure

plt.figure()

plt.subplot(2, 2, 1) # we are defining the subplots here
plt.plot(x, x, color = "red", label = "linear")
plt.legend()
plt.xlim(0,20)
plt.ylim(0,20)

plt.subplot(2, 2, 2)
plt.plot(x, x**2, color = "magenta", label = "parabolic")
plt.legend()
plt.xlim(0,20)
plt.ylim(0,20)

plt.subplot(2, 2, 3)
plt.plot(x, x**3, color = "blue", label = "3rd order")
plt.legend()
plt.xlim(0,20)
plt.ylim(0,20)

plt.subplot(2, 2, 4)
plt.plot(x, x**4, color = "green", label = "4th order")
plt.legend()
plt.xlim(0,20)
plt.ylim(0,20)

plt.show()
# plt.savefig("/content/drive/MyDrive/Colab Notebooks/multipanel.eps")
```

```python
# Method 2: Multi-panel figure using axes objects

fig, ax = plt.subplots(2, 2)

# note how the subplots are defined in a different way

ax[0, 0].plot(x, x, color = "red", label = "linear")
ax[1, 0].plot(x, x**2, color = "magenta", label = "linear")
ax[0, 1].plot(x, x**3, color = "blue", label = "linear")
ax[1, 1].plot(x, x**4, color = "green", label = "linear")

plt.show()
```

```python
# Method 2: Multi-panel figure using axes objects

fig, ax = plt.subplots(nrows = 2, ncols = 2)
ax1, ax2, ax3, ax4 = ax.flatten()

ax1.plot(x, x, color = "red", label = "linear")
ax2.plot(x, x**2, color = "magenta", label = "linear")
ax3.plot(x, x**3, color = "blue", label = "linear")
ax4.plot(x, x**4, color = "green", label = "linear")

plt.show()
```

## Example: Scatter plots (markers)

```python
# Let us generate randon numbers

# Set up a seed
np.random.seed(123)

# Generates randon numbers
new_x = np.random.randn(50)
new_y = np.random.randn(50)

print(new_x[-1])
#print(new_x.shape, new_y.shape)
```

```python
# Open env.
plt.figure(figsize=(4,3))

# Plot
plt.plot(new_x, new_y, label = r"$rand(x)$")

# Title with latex syntax
plt.title("Random numbers")

# Axes titles
plt.xlabel(r"$rand(x)$")
plt.ylabel(r"$rand(y)$")

# Legend is added with loc for location
plt.legend(loc=3)

# Show/Save
plt.show()
```

```python
# Open env.
plt.figure(figsize=(4,3))

# Plot
plt.scatter(new_x, new_y)

# Title with latex syntax
plt.title("Random numbers")

# Axes titles
plt.xlabel(r"$rand(x)$")
plt.ylabel(r"$rand(y)$")

# Legend is added with loc for location
plt.legend(loc=3)

# Show/Save
plt.show()
```

### Example: histograms

```python
# Let's make histograms of new_x and new_y

new_x = np.random.randn(500)
new_y = np.random.randn(500)

#https://matplotlib.org/stable/api/_as_gen/matplotlib.pyplot.hist.html

count, bins, _ = plt.hist(new_x, bins = 20, alpha = 0.6)
count, bins, _ = plt.hist(new_y, bins = 20, alpha = 0.3)
```

```python

count, bins, _ = plt.hist(new_x, bins = 20, alpha = 1., histtype = "step")
count, bins, _ = plt.hist(new_y, bins = 20, alpha = 1., histtype = "step")
```

```python
# Print the objects that plt.hist() returned

print(count.shape)
print(bins.shape)
```

```python
# Uniform Distribution

new_z = np.random.uniform(-3,3,500)

#https://matplotlib.org/stable/api/_as_gen/matplotlib.pyplot.hist.html

# we can save the counts and the bins in variables and use them for other plots or calculations
count, bins, unimportant = plt.hist(new_z, bins = 20, alpha = 0.3)

count, bins, unimportant = plt.hist(new_x, bins = 20, alpha = 0.7)
```

```python
# Uniform Distribution
# We can change the type of the histogram with: histtype

new_z = np.random.uniform(-3,3,500)

#https://matplotlib.org/stable/api/_as_gen/matplotlib.pyplot.hist.html

count1, bins1, unimportant1 = plt.hist(new_z, bins = 20, alpha = 0.3, histtype = "step")

count2, bins2, unimportant2 = plt.hist(new_x, bins = 20, alpha = 0.7, histtype = "step")
```

```python
print(count1.shape, bins1.shape)

print(type(count1))
```

```python
plt.figure(figsize=(4,3))

plt.plot(bins1[1:], count1, linestyle="-", marker = "+", label="uniform")
plt.plot(bins2[1:], count2, linestyle="-", marker = "d", label="normal")

plt.legend()
plt.show()
```
